```python
#getting rid of extra columns
pogo.drop(labels='Attack', axis=1, inplace=True)
pogo.drop(labels='Defense', axis=1, inplace=True)
pogo.drop(labels='Generation', axis=1, inplace=True)
pogo.drop(labels='Legendary', axis=1, inplace=True)
pogo.drop(labels='MaxCP', axis=1, inplace=True)
```

```python
#getting rid of extra variables
pokemonbig.drop(labels='#', axis=1, inplace=True)
pokemonbig.drop(labels='Sp. Atk', axis=1, inplace=True)
pokemonbig.drop(labels='Sp. Def', axis=1, inplace=True)

pokemonbig.drop(labels='Type 1', axis=1, inplace=True)
pokemonbig.drop(labels='Type 2', axis=1, inplace=True)

pokemonbig.drop(labels='Name', axis=1, inplace=True)
```

```python
pokemonbig = pokemonbig.drop_duplicates(subset=['Pokedex'])
```

```python
#erasing extra columns that are repeats or that I don't need
pokemon.drop(labels='type1', axis=1, inplace=True)
pokemon.drop(labels='type2', axis=1, inplace=True)
pokemon.drop(labels='abilities', axis=1, inplace=True)
pokemon.drop(labels='classfication', axis=1, inplace=True)
pokemon.drop(labels='japanese_name', axis=1, inplace=True)
pokemon.drop(labels='name', axis=1, inplace=True)
pokemon.drop(labels='pokedex_number', axis=1, inplace=True)
pokemon.drop(labels='generation', axis=1, inplace=True)
pokemon.drop(labels='is_legendary', axis=1, inplace=True)

pokemon.drop(labels='against_bug', axis=1, inplace=True)
pokemon.drop(labels='against_dark', axis=1, inplace=True)
pokemon.drop(labels='against_dragon', axis=1, inplace=True)
pokemon.drop(labels='against_electric', axis=1, inplace=True)
pokemon.drop(labels='against_fairy', axis=1, inplace=True)
pokemon.drop(labels='against_fight', axis=1, inplace=True)
pokemon.drop(labels='against_fire', axis=1, inplace=True)
pokemon.drop(labels='against_flying', axis=1, inplace=True)
pokemon.drop(labels='against_ghost', axis=1, inplace=True)
pokemon.drop(labels='against_grass', axis=1, inplace=True)
pokemon.drop(labels='against_ground', axis=1, inplace=True)
pokemon.drop(labels='against_ice', axis=1, inplace=True)
pokemon.drop(labels='against_normal', axis=1, inplace=True)
pokemon.drop(labels='against_poison', axis=1, inplace=True)
pokemon.drop(labels='against_psychic', axis=1, inplace=True)
pokemon.drop(labels='against_rock', axis=1, inplace=True)
pokemon.drop(labels='against_steel', axis=1, inplace=True)
pokemon.drop(labels='against_water', axis=1, inplace=True)
pokemon.drop(labels='base_total', axis=1, inplace=True)
pokemon.drop(labels='base_egg_steps', axis=1, inplace=True)
pokemon.drop(labels='sp_attack', axis=1, inplace=True)
pokemon.drop(labels='defense', axis=1, inplace=True)
pokemon.drop(labels='attack', axis=1, inplace=True)
pokemon.drop(labels='sp_defense', axis=1, inplace=True)

pokemon.drop(labels='height_m', axis=1, inplace=True)
pokemon.drop(labels='weight_kg', axis=1, inplace=True)
pokemon.drop(labels='hp', axis=1, inplace=True)
pokemon.drop(labels='capture_rate', axis=1, inplace=True)
```
